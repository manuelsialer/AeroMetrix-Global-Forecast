## Limpieza de Datos

### weather_df: Conversión de la columna timestamp a tipo datetime

```
weather_df['timestamp'] = pd.to_datetime(weather_df['timestamp'])

print('Información del DataFrame de Lecturas Meteorológicas después de la limpieza:')
weather_df.info()
display(weather_df.head())
```

## Verificación y Eliminación de Duplicados

### Duplicados en ocations_df





```
print(f"Número de filas duplicadas en locations_df: {locations_df.duplicated().sum()}")
if locations_df.duplicated().sum() > 0:
    locations_df.drop_duplicates(inplace=True)
    print("Duplicados eliminados de locations_df.")
    print(f"Nuevo número de filas en locations_df: {len(locations_df)}")
```

### Duplicados en weather_df

```
print(f"Número de filas duplicadas en weather_df: {weather_df.duplicated().sum()}")
if weather_df.duplicated().sum() > 0:
    weather_df.drop_duplicates(inplace=True)
    print("Duplicados eliminados de weather_df.")
    print(f"Nuevo número de filas en weather_df: {len(weather_df)}")
```
